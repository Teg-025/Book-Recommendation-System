## Popularity Based Recommender System

```python
ratings.merge(books, on = 'ISBN') ##We merge books on ratings and not the other way around as books have higher rows
```

```python
ratings.merge(books, on = 'ISBN').shape
```

```python
ratings.info() #No changes made here thus we will have to assign a new variable
```

```python
ratings_books = ratings.merge(books, on = 'ISBN')
```

```python
ratings_books.info()
```

```python
ratings_books.head(4)
##Alot of the book titles are same(as many people can rate the same book) thus we will have to group them first
```

```python
number_of_ratings = ratings_books.groupby('Book-Title')['Book-Rating'].count().reset_index()
number_of_ratings.rename(columns= {'Book-Rating': 'Number_Of_Ratings'}, inplace = True)
number_of_ratings

#Counting the no. of ratings(reviews) made
#Can also use pivot_table
#.reset_index used to convert series into dataframe
```

```python
average_ratings = ratings_books.groupby('Book-Title')['Book-Rating'].mean().reset_index()
average_ratings.rename(columns= {'Book-Rating': 'Average_Ratings'}, inplace = True)
average_ratings
```

```python
popular_books = number_of_ratings.merge(average_ratings, on='Book-Title')
popular_books

#We just modified are dataset for containing ratings and the no. of reviews
```

```python
#Now we can select only those books which have more than 40 reviews

popular_books = popular_books[popular_books['Number_Of_Ratings']>=100]
popular_books
```

```python
popular_books.sort_values('Average_Ratings', ascending=False)
```

```python
#Now that we have sorted out those books with less reviews we will now want to merge it with images and author name

popular_books.merge(books, on='Book-Title').drop_duplicates('Book-Title')

#If we don't drop duplicates all those books having same title but different isbn for paperback and translated etc will also show 
```

```python
popular_books= popular_books.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author', 'Number_Of_Ratings', 'Average_Ratings','Image-URL-S']]
```

## Collabirative based reccomendation

```python
#What we are going to do is filter out those users first that have not made more than 200 ratings
##for our recommendation system
ratings_books.head(4)
```

```python
ratings_books.isnull().sum()
```

```python
books_ratings_filter1 = ratings_books.groupby('User-ID')['Book-Rating'].count()>200

user_filter = books_ratings_filter1[books_ratings_filter1].index
```

```python
filtered_rating = ratings_books[ratings_books['User-ID'].isin(user_filter)]
```

```python
filtered_rating.head(4)
```

```python
#Then we filter out books with less than 50 ratings

books_ratings_filter2 = filtered_rating.groupby('Book-Title')['Book-Rating'].count()>50
famous_books_filter = books_ratings_filter2[books_ratings_filter2].index
```

```python
famous_books_filter
```

```python
final_filter = filtered_rating[filtered_rating['Book-Title'].isin(famous_books_filter)]
```

```python
final_filter.head()
```

```python
final_filter.drop_duplicates()
```

```python
#This only contains book-title and user-id with no. of ratings>200 for users who have rated>50 books
#pt refers to Pivot Table
pt = final_filter.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
pt.head(4)
```

```python
pt.fillna(0,inplace=True)
```

```python
pt.head(4)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Calculated euclidian distance from one book to all others
cosine_similarity(pt)
```

```python
cosine_similarity(pt).shape
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
def recommend(book_name):
    #Fetch Index
    index = np.where(pt.index == book_name)[0][0]
    
    #Storing this in list along with corresponding index whose similarity
    #score it is
    similar_list = list(enumerate(similarity_scores[index]))
    
    #Performing sorting in reverse order so that the one having highest
    #similarity is the first and then only taking the top 6
    similar_items = sorted(similar_list, key=lambda x:x[1], reverse = True)[1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Message in a Bottle')
```

```python
recommend('1984')
```
